```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Download necessary NLTK data
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Define the dataset (as created earlier)
data = {
    'Product_ID': ['001', '002', '003', '004', '005', '006', '007', '008', '009', '010',
                   '011', '012', '013', '014', '015', '016', '017', '018', '019', '020',
                   '021', '022', '023', '024', '025', '026', '027', '028', '029', '030'],
    'Product_Name': ['AlphaPhone X1', 'AlphaPhone X2', 'BetaTab Pro', 'BetaTab Mini', 'GammaWatch S1',
                     'GammaWatch S2', 'DeltaLaptop 14', 'DeltaLaptop 16', 'EchoEarbuds Pro', 'EchoEarbuds Lite',
                     'AlphaPhone X3', 'BetaTab Max', 'GammaWatch S3', 'DeltaLaptop 13', 'EchoSpeaker Pro',
                     'EchoSpeaker Lite', 'AlphaPhone X Lite', 'BetaTab Lite', 'GammaWatch Lite', 'DeltaLaptop Pro',
                     'EchoEarbuds Max', 'AlphaPhone SE', 'BetaTab Ultra', 'GammaWatch Max', 'DeltaLaptop Ultra',
                     'EchoSpeaker Max', 'AlphaPhone XL', 'BetaTab Plus', 'GammaWatch X', 'EchoSpeaker Pro+'],
    'Category': ['Smartphone', 'Smartphone', 'Tablet', 'Tablet', 'Smartwatch',
                 'Smartwatch', 'Laptop', 'Laptop', 'Earbuds', 'Earbuds',
                 'Smartphone', 'Tablet', 'Smartwatch', 'Laptop', 'Speaker',
                 'Speaker', 'Smartphone', 'Tablet', 'Smartwatch', 'Laptop',
                 'Earbuds', 'Smartphone', 'Tablet', 'Smartwatch', 'Laptop',
                 'Speaker', 'Smartphone', 'Tablet', 'Smartwatch', 'Speaker'],
    'Price': [699, 799, 499, 349, 199,
              249, 999, 1299, 149, 99,
              899, 649, 299, 899, 199,
              149, 599, 299, 179, 1499,
              199, 499, 799, 349, 1699,
              249, 1099, 549, 399, 299],
    'Units_Sold': [2500, 2200, 1800, 1500, 3000,
                   2800, 1200, 1000, 3200, 3400,
                   2000, 1700, 2600, 1400, 2200,
                   2400, 2700, 1600, 3100, 900,
                   3000, 2300, 1300, 2400, 800,
                   2100, 1900, 1400, 2200, 2000],
    'Rating': [4.5, 4.6, 4.3, 4.1, 4.4,
               4.6, 4.7, 4.8, 4.2, 4.1,
               4.7, 4.5, 4.6, 4.4, 4.3,
               4.1, 4.3, 4.0, 4.2, 4.9,
               4.4, 4.3, 4.6, 4.7, 4.9,
               4.5, 4.8, 4.4, 4.8, 4.6]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Function to preprocess and tokenize the user query
def preprocess_query(query):
    # Tokenize the query
    tokens = word_tokenize(query.lower())

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word.isalnum() and word not in stop_words]

    # Lemmatize the tokens
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]

    return lemmatized_tokens

# Function to calculate similarity score between query and dataset row
def calculate_similarity(row, query_tokens):
    # Concatenate all string values of the row
    row_text = ' '.join(str(value).lower() for value in row)

    # Tokenize row text
    row_tokens = word_tokenize(row_text)

    # Count matches between query tokens and row tokens
    match_count = sum(1 for token in query_tokens if token in row_tokens)

    return match_count

# Function to find the most relevant row based on user query
def find_most_similar_row(df, query):
    # Preprocess the query
    query_tokens = preprocess_query(query)

    # Calculate similarity for each row
    df['similarity_score'] = df.apply(lambda row: calculate_similarity(row, query_tokens), axis=1)

    # Find the row with the highest similarity score
    most_similar_row = df.loc[df['similarity_score'].idxmax()]

    return most_similar_row

# Example usage
query = input("Ask a question about the product (e.g., 'Which tablet costs 499?'): ")
result = find_most_similar_row(df, query)

# Print the result
print("\nMost relevant product:")
print(result)
```
